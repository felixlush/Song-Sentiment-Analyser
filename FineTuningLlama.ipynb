```python
import os
import torch
from datasets import load_dataset, Dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    TrainingArguments,
    pipeline,
    logging,
)

from peft import LoraConfig, PeftModel, get_peft_model
from trl import SFTTrainer, SFTConfig
```

```python
model_name = 'NousResearch/Llama-2-7b-chat-hf'
dataset_name = 'GammaKing2000/NLP_assignment_3'
fine_tuned_model = 'Llama2MusicSentimentBot'

# QLoRA parameters
lora_r = 64
lora_alpha = 16
lora_dropout = 0.1

# bitsandbytes parameters
use_4bit = True
bnb_4bit_compute_dtype = "float16"
bnb_4bit_quant_type = "nf4"
use_nested_quant = False

# TrainingArguments parameters
output_dir = "./results"
num_train_epochs = 1
fp16 = False
bf16 = True
per_device_train_batch_size = 1
per_device_eval_batch_size = 1
gradient_accumulation_steps = 1
gradient_checkpointing = True
max_grad_norm = 0.3
learning_rate = 2e-4
weight_decay = 0.001
optim = "paged_adamw_32bit"
lr_scheduler_type = "cosine"
max_steps = -1
warmup_ratio = 0.03
group_by_length = True# Group sequences into batches with same length. Saves memory and speeds up training.
save_steps = 0
logging_steps = 25

# SFT parameters
max_seq_length = 512
packing = False
device_map = {"": 0}
```

```python
dataset = load_dataset(dataset_name, split='train')
```

```python
compute_dtype = getattr(torch, bnb_4bit_compute_dtype)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=use_4bit,
    bnb_4bit_quant_type=bnb_4bit_quant_type,
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=use_nested_quant,
)

# Load base model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map=device_map
)
model.config.use_cache = False
model.config.pretraining_tp = 1

# Load LLaMA tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'right' # Fix weird overflow issue with fp16 training

# Load LoRA configuration
peft_config = LoraConfig(
    lora_alpha=lora_alpha,
    lora_dropout=lora_dropout,
    r=lora_r,
    bias='none',
    task_type='CAUSAL_LM',
)

# Set training parameters
training_arguments = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=num_train_epochs,
    per_device_train_batch_size=per_device_train_batch_size,
    gradient_accumulation_steps=gradient_accumulation_steps,
    optim=optim,
    save_steps=save_steps,
    logging_steps=logging_steps,
    learning_rate=learning_rate,
    weight_decay=weight_decay,
    fp16=fp16,
    bf16=bf16,
    max_grad_norm=max_grad_norm,
    max_steps=max_steps,
    warmup_ratio=warmup_ratio,
    group_by_length=group_by_length,
    lr_scheduler_type=lr_scheduler_type,
    report_to='tensorboard'
)
```

```python
final_prompts = []
for data in dataset:
  text_entry = f"<s>[INST] <<SYS>>\n{data['system_prompt']}<</SYS>>\n\n{data['question']} [/INST] {data['response']} </s>"
  final_prompts.append(text_entry)
```

```python
import pandas as pd
final_inputs = pd.DataFrame({'text':final_prompts})
```

```python
import os
os.environ["TORCH_CUDA_ARCH_LIST"] = "80"  # For CUDA 12.4
os.environ["CUDA_VISIBLE_DEVICES"] = "0"  # If you have a single GPU
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True" # Add this line
```

```python
model = get_peft_model(model, peft_config)
print("PEFT adapter applied to the model. Trainable parameters:")
model.print_trainable_parameters()

num_rows = len(final_inputs)
for i in range(0, 20):
    start_index = int(num_rows * 0.05 * i)
    end_index = int(num_rows * 0.05 * (i + 1))

    new_dataset_data = final_inputs[start_index:end_index]
    new_dataset = Dataset.from_dict(new_dataset_data)

    print(f"\n--- Iteration {i+1}/20: Training on dataset chunk from index {start_index} to {end_index} ---")

    sft_config = SFTConfig(
        output_dir=output_dir,
        num_train_epochs=num_train_epochs,
        per_device_train_batch_size=per_device_train_batch_size,
        gradient_accumulation_steps=gradient_accumulation_steps,
        optim=optim,
        save_steps=save_steps,
        logging_steps=logging_steps,
        learning_rate=learning_rate,
        weight_decay=weight_decay,
        fp16=fp16,
        bf16=bf16,
        max_grad_norm=max_grad_norm,
        max_steps=max_steps,
        warmup_ratio=warmup_ratio,
        group_by_length=group_by_length,
        lr_scheduler_type=lr_scheduler_type,
        report_to="tensorboard",
        max_seq_length=max_seq_length,
        packing=packing,
    )

    trainer = SFTTrainer(
        model=model,
        train_dataset=new_dataset,
        peft_config=peft_config,
        args=sft_config,
        processing_class=tokenizer,
    )

    print(f"Starting training for chunk {i+1}...")
    trainer.train()
    print(f"Finished training for chunk {i+1}.")

print("\n--- All training chunks processed ---")
```

```python
try:
    print("\nMerging adapter with the base model...")
    merged_model = model.merge_and_unload()
    print("Adapter merged.")

    merged_model_path = os.path.join(output_dir, "final_merged_model")
    if not os.path.exists(merged_model_path):
        os.makedirs(merged_model_path)

    merged_model.save_pretrained(merged_model_path)
    tokenizer.save_pretrained(merged_model_path)
    print(f"Final merged model and tokenizer saved to: {merged_model_path}")

except Exception as e:
    print(f"Error during model merging or saving merged model: {e}")
    print("Skipping merged model saving. The adapter itself was saved successfully.")
```

```python
def generate_text(prompt, max_length=200):
    print("\nGenerating response.....\n\n")
    system_prompt = """[SYS]you are an AI assistant. User will give you some details about a song including artist name, song name, and the song lyrics, since you are very well versed in recognising the sentiment of these song lyrics you have to tell the user what is the sentiment of that song.
give the answer in this format, do not give anything extra other than this sentence:
The song '[song name]' by '[artist]' has a sentiment of being '[sentiment]'[/SYS]
"""
    fPrompt = system_prompt + f"<s>[INST] {prompt} [/INST]"
    inputs = tokenizer_for_inference_merged(fPrompt, return_tensors="pt").to(loaded_merged_model.device)
    outputs = loaded_merged_model.generate(**inputs, max_new_tokens=100)
    ans = tokenizer_for_inference_merged.decode(outputs[0], skip_special_tokens=True)
    print(ans.split("[/INST]")[-1].split(".")[0])
```

```python
import zipfile
import os

def zip_folder(folder_path, zip_name):
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # Walk through all files in the directory
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # Create the full file path
                file_path = os.path.join(root, file)
                # Calculate path for file in zip
                arcname = os.path.relpath(file_path, folder_path)
                # Add file to zip
                zipf.write(file_path, arcname)

# Example usage
zip_folder('Song-Sentiment-Analyser/datasets/results/final_merged_model', 'my_archive.zip')
```
